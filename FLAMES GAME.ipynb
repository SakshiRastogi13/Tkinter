```python
from tkinter import *
from tkinter import messagebox
root=Tk()
root.title("FLAMES GAME")
root.geometry("700x400")
root.configure(background="Light Pink")
def submit():
    n1=E1.get()
    n2=E2.get()
    NL1=list(n1.upper())
    NL2=list(n2.upper())
    for l in NL1:
        print("Checking ",l)
        if l in NL2:
            print("Found ",l)
            i=NL1.index(l)
            NL1[i]="_"
            i=NL2.index(l)
            NL2[i]="_"
    print(NL1)
    print(NL2)
    name=NL1+NL2
    print(name)
    count=0
    for n in name:
        if(n!='_' and n!=' '):
            count=count+1
    print(count)
    f='FLAMES'
    F1=list(f)
    print(F1)
    index=0
    while(len(F1)>1):
        for j in range(count):
            index=index+1
            if(index>len(F1)):
                index=1
        F1.pop(index-1)
        index=index-1
    L4["text"]=F1

def Quit():
    q=messagebox.askyesno("Do you want to quit?")
    if(q>0):
        root.destroy()
        return
    
def Clear():
    E1.delete(0,END)
    E2.delete(0,END)
    
            
L1=Label(root,text="Player 1 Name:-",font=('arial',25,'bold'),fg='Blue')
L1.grid(row=0,column=0)
L2=Label(root,text="Player 2 Name:-",font=('arial',25,'bold'),fg='Blue')
L2.grid(row=1,column=0)
E1=Entry(root,width=40,border=5)
E1.grid(row=0,column=4)
E2=Entry(root,width=40,border=5)
E2.grid(row=1,column=4)
B1=Button(root,text="Submit",width=15,height=2,bg="Yellow",fg="Red",font=('arial',10,'bold'),command=submit)
B1.grid(row=3,column=3)
L3=Label(root,text="Relationship Status:-",font=('arial',20,'bold'),fg='Green')
L3.grid(row=6,column=0)
E3=Entry(root,width=40,border=5)
E3.grid(row=6,column=4)
B2=Button(root,text="Clear",width=15,height=2,bg="Yellow",fg="Red",font=('arial',10,'bold'),command=Clear)
B2.grid(row=8,column=3)
B3=Button(root,text="Quit",width=15,height=2,bg="Yellow",fg="Red",font=('arial',10,'bold'),command=Quit)
B3.grid(row=8,column=4)
L4=Label(root,text="\nResult\n",font=('arial',20,'bold'),fg="White",bg="Black")
L4.grid(row=13,column=2)
root.mainloop()
```
