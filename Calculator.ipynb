```python
from tkinter import *
import tkinter as tk
root=tk.Tk()
root.title("CALCULATOR")
root.geometry('600x600')
a=Entry(root,width=50,borderwidth=10)
a.grid(row=0,column=0,columnspan=3,padx=40,pady=40)
def button_click(number):
    b=a.get()
    a.delete(0,END)
    a.insert(0,str(b)+str(number))
def button_clear():
    a.delete(0,END)
def button_add():
    f_num=a.get()
    global first_number
    global math
    math='addition'
    first_number=int(f_num)
    a.delete(0,END)
def button_multiply():
    f_num=a.get()
    global first_number
    global math
    math='multiplication'
    first_number=int(f_num)
    a.delete(0,END)
def button_subtract():
    f_num=a.get()
    global first_number
    global math
    math='substraction'
    first_number=int(f_num)
    a.delete(0,END)

def button_divide():
    f_num=a.get()
    global first_number
    global math
    math='division'
    first_number=int(f_num)
    a.delete(0,END)

def button_equal():
    s_num=a.get()
    second_number=int(s_num)
    a.delete(0,END)
    if(math=='addition'):
        a.insert(0,first_number + second_number)
    elif(math=='multiplication'):
        a.insert(0,first_number * second_number)
    elif(math=='substraction'):
        a.insert(0,first_number - second_number)
    elif(math=='division'):
        a.insert(0,first_number / second_number)


button1=Button(root,text="1",padx=50,pady=30,command=lambda:button_click(1))
button2=Button(root,text="2",padx=50,pady=30,command=lambda:button_click(2))
button3=Button(root,text="3",padx=50,pady=30,command=lambda:button_click(3))
button4=Button(root,text="4",padx=50,pady=30,command=lambda:button_click(4))
button5=Button(root,text="5",padx=50,pady=30,command=lambda:button_click(5))
button6=Button(root,text="6",padx=50,pady=30,command=lambda:button_click(6))
button7=Button(root,text="7",padx=50,pady=30,command=lambda:button_click(7))
button8=Button(root,text="8",padx=50,pady=30,command=lambda:button_click(8))
button9=Button(root,text="9",padx=50,pady=30,command=lambda:button_click(9))
button0=Button(root,text="0",padx=50,pady=30,command=lambda:button_click(0))
button_add=Button(root,text="+",padx=50,pady=30,command=button_add)
button_multiply=Button(root,text="*",padx=50,pady=30,command=button_multiply)
button_subtract=Button(root,text="-",padx=50,pady=30,command=button_subtract)
button_divide=Button(root,text="/",padx=50,pady=30,command=button_divide)
button_clear=Button(root,text="CLR",padx=45,pady=30,command=button_clear)
button_equal=Button(root,text="=",padx=50,pady=30,command=button_equal)




button1.grid(row=1,column=0)
button2.grid(row=1,column=1)
button3.grid(row=1,column=2)
button4.grid(row=2,column=0)
button5.grid(row=2,column=1)
button6.grid(row=2,column=2)
button7.grid(row=3,column=0)
button8.grid(row=3,column=1)
button9.grid(row=3,column=2)
button0.grid(row=4,column=0)
button_add.grid(row=4,column=1)
button_multiply.grid(row=5,column=1)
button_subtract.grid(row=5,column=2)
button_divide.grid(row=5,column=3)
button_clear.grid(row=4,column=2)
button_equal.grid(row=4,column=3)



root.mainloop()
```
